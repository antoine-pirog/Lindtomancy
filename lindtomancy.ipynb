```python
import utils
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data_file = "chocolate.json"
objdata = utils.load_chocodata(data_file)
df = utils.build_chocodataframe(objdata)
df.head()
```

```python
h, ax = plt.subplots(figsize=(5,9))
df.sum(axis=1).sort_values().plot(ax=ax, kind="barh", xlabel="Occurences")
ax.xaxis.grid(True, which='major', ls=':')
plt.show()
```
